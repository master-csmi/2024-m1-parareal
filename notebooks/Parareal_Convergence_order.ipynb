# Convergence Order of Parareal Algorithm for Different Solver Combinations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read data from file
df = pd.read_csv("./python/data/data_cv_order_2024-07-16_15-55-38.txt", sep=' ', header=None, names=['Function', 'Coarse', 'Fine', 'Order'])

# Filter data for each function
df_f = df[df['Function'] == 'f']
df_z = df[df['Function'] == 'z']
df_h = df[df['Function'] == 'h']

# Function to plot results
def plot_results(data, title):
    plt.figure(figsize=(12, 6))
    sns.barplot(data=data, x='Coarse', y='Order', hue='Fine', errorbar=None)
    plt.title(title)
    plt.ylabel('Convergence Order')
    plt.xticks(rotation=45)
    plt.show()
```

1 - 
$$
\frac{dy}{dt}=-ty^2
$$

```python
plot_results(df_f, 'Convergence Order of Parareal Algorithm for Function f')
```

2 - 
$$
\frac{dy}{dt}=-y + cos(t)
$$

```python
plot_results(df_z, 'Convergence Order of Parareal Algorithm for Function z')
```

3 -
$$
\frac{dy}{dt}= -2y
$$

```python
plot_results(df_h, 'Convergence Order of Parareal Algorithm for Function h')
```
